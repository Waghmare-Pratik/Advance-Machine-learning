Implementation of Apriori Algorithm

```python
import numpy as np
import matplotlib as mtp
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
df = pd.read_csv("Aprior.csv")
df.tail()
```

```python
df.shape
```

```python
transactions=[]
for i in range(0, 9):
  transactions.append([str(df.values[i,j]) for j in range(0,5)])
```

```python
pip install apyori
```

```python
 
from apyori import apriori
rules = list(apriori(transactions=transactions, min_support=0.003, min_confidence=0.2))
results=list(rules)
results
```

```python
for item in results:
    pair = item[0]
    if len(pair) > 1:
        items = [x for x in pair]
        print("Rule:" + items[0] + "->" + items[1])
    else:
        print("Rule:" + str(pair))

    print("Support:" + str(item[1]))
    print("Confidence:" + str(item[2][0][2]))
    print("Lift:" + str(item[2][0][3]))
    print("================ ===============")
```
